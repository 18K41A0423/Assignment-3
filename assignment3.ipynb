```
#2. Draw correlation graph for mobile price with all attributes.
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.max_open_warning':0})
df=pd.read_csv("mobile_cleaned-1549119762886_-452616671.csv")
for col in df.columns:
  if(col!='price'):
    df.plot(x='price',y=col)
    plt.ylabel(col)
    plt.title("Relation between price and %s"%col)
```

```
#3. Compute outer product of 2 vectors.
import numpy as np
p=[[1,0],[0,1]]
q=[[1,2],[3,4]]
print("Original matrix:",p,q)
res=np.outer(p,q)
print("Outer product of two vectors :\n",res)
```

```
#4.  Write a NumPy program to compute the condition number of a given matrix.
#Original matrix:[[1 2][3 4]]8
#Condition number of the said matrix:14.93303437375.
import numpy as np
from numpy import linalg
m=np.array([[1,2],[3,4]])
print("Original Matrix:\n",m)
res=np.linalg.cond(m)
print("Condition number of the matrix :",res)
```

```
#5. write a NumPy program to compute the sum of the diagonal element of a given array(matrix).
import numpy as np
x=int(input("Number of element is in the matrix:"))
y=int(input("Enter order of matrix:"))
z=int(input("Enter order of matrix:"))
m=np.arange(x).reshape(y,z)
print(m)
res=np.trace(m)
print("Sum of diagonal elements:",res)
```
